```python
INF = 10**18

def solve(lines):
    query = lines[-1] # две вершины
    paths = lines[:-1] # пути

    # извлекаем символы a, b
    qchars = [c for c in query if not c.isspace()]
    a = qchars[0]
    b = qchars[1]

    # у вершины 0 имя '/'
    children = [dict()]
    name = ['/']

    def new_node(ch):
        children.append(dict())
        name.append(ch) # имя вершины
        return len(name) - 1 # возвращаем индекс вешины

    # построение дерева по путям
    for path in paths:
        cur = 0
        for ch in path:
            if ch == '/':
                continue
            if ch not in children[cur]:
                children[cur][ch] = new_node(ch)
            cur = children[cur][ch]

    n = len(name) # число всех вершин дерева

    order = []
    stack = [0]
    while stack:
        v = stack.pop()
        order.append(v)
        for to in children[v].values():
            stack.append(to)
    order.reverse() # массив

    minA = [INF] * n # мин расстояние от v до a
    minB = [INF] * n
    maxA = [-INF] * n
    maxB = [-INF] * n

    # ответы
    ans_min = INF
    ans_max = -INF

    # обход снизу вверх
    for v in order:
        # если вершина v имеет имя a
        if name[v] == a:
            minA[v] = 0
            maxA[v] = 0
        # если вершина v имеет имя b
        if name[v] == b:
            minB[v] = 0
            maxB[v] = 0

        # если a=b
        if minA[v] == 0 and minB[v] == 0:
            ans_min = min(ans_min, 0)
            ans_max = max(ans_max, 0)

        curMinA = minA[v] # расстояние от v до ближайшей вершины a
        curMinB = minB[v]
        curMaxA = maxA[v]
        curMaxB = maxB[v]

        for to in children[v].values():
            if curMinA < INF and minB[to] < INF:
                ans_min = min(ans_min, curMinA + 1 + minB[to])
            if curMinB < INF and minA[to] < INF:
                ans_min = min(ans_min, curMinB + 1 + minA[to])

            if curMaxA > -INF and maxB[to] > -INF:
                ans_max = max(ans_max, curMaxA + 1 + maxB[to])
            if curMaxB > -INF and maxA[to] > -INF:
                ans_max = max(ans_max, curMaxB + 1 + maxA[to])

            if minA[to] < INF:
                minA[v] = min(minA[v], 1 + minA[to])
                curMinA = min(curMinA, 1 + minA[to])

            if minB[to] < INF:
                minB[v] = min(minB[v], 1 + minB[to])
                curMinB = min(curMinB, 1 + minB[to])

            if maxA[to] > -INF:
                maxA[v] = max(maxA[v], 1 + maxA[to])
                curMaxA = max(curMaxA, 1 + maxA[to])

            if maxB[to] > -INF:
                maxB[v] = max(maxB[v], 1 + maxB[to])
                curMaxB = max(curMaxB, 1 + maxB[to])

    print(ans_min, ans_max)
```

```python
data = [
    "/a",
    "/b",
    "a b"
]

solve(data)
```

```python
data = [
    "/a",
    "/b/a",
    "a b"
]

solve(data)
```

```python
data = [
    "/a/b/c",
    "/a/c",
    "d/b/x",
    "d/c",
    "x c"
]

solve(data)
```

```python
data = [
    "/a",
    "/b/c/c/a",
    "/b/c/a",
    "/b/b/c/b",
    "/b/b/c/c",
    "b/b",
    "/b/b/c",
    "a b"
]

solve(data)
```
