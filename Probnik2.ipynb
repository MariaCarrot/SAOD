```python
from collections import defaultdict

def solve(lines):
    # у вершины 0 имя '/'
    children = [dict()]
    name = ['/']

    def new_node(ch):
        children.append(dict())
        name.append(ch) # имя вершины
        return len(name) - 1 # возвращаем индекс вешины

    for path in lines:
        path = path.strip()
        if not path:
            continue
        if path[0] != '/':
            raise ValueError(f"Некорректный путь: {path}")

        cur = 0
        for ch in path:
            if ch == '/':
                continue
            if ch not in children[cur]:
                children[cur][ch] = new_node(ch)
            cur = children[cur][ch]

    n = len(name)

    # порядок снизу вверх
    order = []
    stack = [0]
    while stack:
        v = stack.pop()
        order.append(v)
        for to in children[v].values():
            stack.append(to)
    order.reverse()

    # Канонизация поддеревьев
    signature_to_id = {}
    type_id = [0] * n
    freq = defaultdict(int) # сколько раз встретился каждый тип

    next_id = 0

    for v in order:
        child_info = []
        for ch, to in children[v].items():
            child_info.append((ch, type_id[to]))

        child_info.sort()
        signature = (name[v], tuple(child_info))

        if signature not in signature_to_id:
            signature_to_id[signature] = next_id
            next_id += 1

        type_id[v] = signature_to_id[signature]
        freq[type_id[v]] += 1

    # Считаем, сколько разных типов встретились >= 2 раз
    answer = 0
    for t, count in freq.items():
        if count >= 2:
            answer += 1

    print(answer)
```

```python
data = [
    "/a",
    "/b",
    "/c"
]

solve(data)
```
